```python
#import essential modules and libraries
#ntscrapper -- unofficial library to scrape twitter data, but this library is still in developement.

from ntscraper import Nitter
import pandas as pd 
import datetime
import ntscraper
```

```python
scraper= Nitter(log_level =1)

#created an object (scraper) for "NITTER" module to access/scrape functionalities from twitter 
```

```python
#IF WE WANT TO SCRAPE INFORMATION ABOUT SOME SPECIFIC USER/TWEETS, WE CAN USE FUNCTION "GET_TWEETS" FROM NITTER MODULE

tweets= scraper.get_tweets("elonmusk", mode="user", number=1000)
```

```python
print(tweets)
```

```python
#WE HAVE ALREADY GOT 100 TWEETS, BUT THEY ARE HIDDEN CAUSE OF NITTER INSTANCE MIGHT HAVE NOT CONTENT/DATA WE ARE TRYING TO SCRAPE NOW.

#PPRINT -> pprint is a module in Python's standard library that provides functionality to pretty-print data structures in a more human-readable format. It's commonly used for formatting complex data structures such as dictionaries, lists, and nested objects.
from pprint import pprint
```

```python
pprint(tweets)
```

```python
#FUNCTION (GET_PROFILE_INFO)  TO GET INFORMATION ABOUT A SPECIFIC USER FROM THEIR PROFILE

elon_info=scraper.get_profile_info(username= "elonmusk")
```

```python
#here is the information about specific user, here elon musk from elon_info
pprint(elon_info)
```

```python
#TO GET ONLY TOP TWEET FROM ELON

tweets['tweets'][0]
```

```python
#NOW CONVERT THE SCATTERED DATA INTO DICTIONARY AND FROM DICT TO CSV/DATAFRAME

data={
       'link' :[], 
       'text' :[], 
       'user' :[],
       'likes':[],
       'comments':[],
       'retweets':[],
       'external-link':[],
       'quotes': []
}

for tweet in tweets['tweets']:
    data['link'].append(tweet['link'])
    data['text'].append(tweet['text'])
    data['user'].append(tweet['user']['name'])
    data['likes'].append(tweet['stats']['likes'])
    data['comments'].append(tweet['stats']['comments'])
    data['quotes'].append(tweet['stats']['quotes'])
    data['retweets'].append(tweet['stats']['retweets'])
    data['external-link'].append(tweet['external-link'])                      
                        
```

```python
pprint(data)
```

```python
#PANDAS TO CONVERT OUR DATA INTO CSV

df=pd.DataFrame(data)
df.head()
```

```python
#CONVERT IT INTO CSV

df.to_csv("elon_X_tweets")
```
